### Converting JSON Columns to Child Dataframes

```python
import json  # for parsing JSON strings
```

#### Function to Convert JSON Column to Child DataFrame

```python
def json_to_df(df, json_col, movie_id_col='movie_id'):
    """
    Converts a JSON column to a separate DataFrame with movie_id.
    """
    rows = []  # creating a list to hold flattened rows
    
    for idx, row in df.iterrows():  # iterating through each row of parent df
        movie_id = row[movie_id_col]  # getting movie_id
        json_data = row[json_col]  # accessing JSON string
        if pd.isna(json_data):  # checking for NaN
            continue
        try:
            items = json.loads(json_data)  # converting JSON string to Python object
            for item in items:  # iterating over each item in the list
                item[movie_id_col] = movie_id  # adding movie_id to child row
                rows.append(item)  # appending to rows
        except json.JSONDecodeError:
            continue
    
    return pd.DataFrame(rows)  # returning the child dataframe
```

#### Applying Function to Each JSON Column

```python
# creating separate child dataframes from JSON columns

genres_df = json_to_df(df, 'genres') 
keywords_df = json_to_df(df, 'keywords')  
production_companies_df = json_to_df(df, 'production_companies')  
production_countries_df = json_to_df(df, 'production_countries')  
cast_df = json_to_df(df, 'cast')  
crew_df = json_to_df(df, 'crew') 

# Each child DataFrame now has all original attributes + movie_id
```

## Data Preprocessing

### Converting non-numeric labeled columns into categorical types

```python
print('Before Conversion\n', m_df.dtypes)
print('-'*30)

def non_num_to_cat(m_df):
    """
    Converts all object dtype columns in df to category dtype.
    Returns a new DataFrame with conversions applied.
    """
    # Finding all object dtype columns
    non_num_cols = m_df.select_dtypes(include=['object']).columns
    
    # Converting them to category
    for col in non_num_cols:
        m_df[col] = m_df[col].astype('category')
    return m_df
m_df = non_num_to_cat(m_df)
print('After Conversion\n', m_df.dtypes)
```
